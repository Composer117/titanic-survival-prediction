```python
titanic_df = titanic_df.drop(['PassengerId', 'Name', 'Ticket', 'Cabin'], axis=1)
titanic_test = titanic_test.drop(['PassengerId', 'Name', 'Ticket', 'Cabin'], axis=1)
```

```python
titanic_df_final = pd.get_dummies(titanic_df)
titanic_test_final = pd.get_dummies(titanic_test)
```

```python
def create_missing_indicator(df):
    if np.isnan(df['Age']):
        return 1
    else:
        return 0

titanic_df_final['missing_age'] = titanic_df_final.apply(create_missing_indicator, axis=1)
titanic_test_final['missing_age'] = titanic_test_final.apply(create_missing_indicator, axis=1)
titanic_df_final.loc[titanic_df_final['Age'].isna(), 'Age'] = titanic_df_final['Age'].mean()
titanic_test_final.loc[titanic_test_final['Age'].isna(), 'Age'] = titanic_df_final['Age'].mean()
```

```python
train_X = titanic_df_final.drop('Survived', axis=1)
train_y = titanic_df_final['Survived']
```

```python
train_X
```

```python
from sklearn.model_selection import train_test_split

X_train, X_test, y_train, y_test = train_test_split(train_X, train_y, train_size=0.33, random_state=42)
```

```python
from sklearn.model_selection import GridSearchCV
from sklearn.ensemble import RandomForestClassifier

parameters = {'n_estimators':[10, 25, 50, 100], 'max_features': list(range(1, X_train.shape[1]+1))}
rf = RandomForestClassifier()

clf_model = GridSearchCV(rf, parameters)
```

```python
clf_model.fit(X_train, y_train)
```

```python
from sklearn.metrics import accuracy_score

accuracy_score(y_test, clf_model.predict(X_test))
```
